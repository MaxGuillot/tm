The procedural python solution
-----------

The idea of the following code is to parse every character within the json file. Every specific character defined in the special dictionnary will trigger the adequated modification. Any other character will be printed without further changes.

```python
fic = open("tweet0.json", "r")
special = {
    "{":"open",
    "(":"open",
    "[":"open",
    "}":"close",
    ")":"close",
    "]":"close",
    ",":"lf",
    '"':"str",
    "'":"str"
}

string = False

output = ""

indent = 0

for ligne in fic.readlines():
    for char in ligne:
        if char in special.keys():
            instr = special[char]
            if instr == "open":
                output += char + "\n"
                indent += 1
                for i in range(indent):
                    output += "\t"
            elif instr == "close":
                output += "\n"
                indent -= 1
                for i in range(indent):
                    output += "\t"
                output += char
            elif instr == "lf":
                output += char + "\n"
                for i in range(indent):
                    output += "\t"
            elif instr == "str":
                output += char
                if string:
                    string = False
                else:
                    string = True
        else:
            if string or char != " ":
                output += char
print(output)
fic.close()
```
